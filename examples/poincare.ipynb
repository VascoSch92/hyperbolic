```python
def hyp_poly_edge_construct(p, q):
    pi, pi2 = math.pi, math.pi*2
    th = pi2/q
    phi = pi2/p
    ang1 = pi-phi/2-th/2-pi/2
    ang2 = th/2 + pi/2
    a = math.sin(ang2)/math.sin(ang1)
    b = math.sin(phi/2)/math.sin(ang1)
    r_p = math.sqrt(1/(a**2-b**2))
    r_c = a*r_p
    r_from_c = b*r_p
    #return r_c, r_from_c
    t1 = pi - math.asin(r_c / (r_from_c / math.sin(phi/2)))
    t2 = pi - t1 - phi/2
    r = math.sin(t2) * (r_from_c / math.sin(phi/2))
    return r
```

```python
class TreeNode:
    def __init__(self, point, children=None, parent=None):
        self.point = point
        self.children = [] if children is None else children
        self.parent = parent
    def add_child(self, child):
        child.parent = self
        self.children.append(child)
    def transform(self, trans):
        self.point = trans.apply_to_point(self.point)
        for child in self.children:
            child.transform(trans)
    def transformed(self, trans, new_parent):
        new_point = trans.apply_to_point(self.point)
        new_node = TreeNode(new_point, None, new_parent)
        new_node.children = [child.transformed(trans, new_node) for child in self.children]
        return new_node
    @staticmethod
    def make_linked_list(points):
        head = None
        prev = None
        for p in points:
            nxt = TreeNode(p)
            if prev is None:
                head = nxt
            else:
                prev.add_child(nxt)
            prev = nxt
        return head
    def calc_polygon_edges_and_points(self, cap=False, cap_start=False, reverse_caps=False,
                                      reverse_start_cap=False):
        edges, points = [], []
        for child in self.children:
            edge_first = Line.from_points(*self.point, *child.point, segment=True)
            edge_last = edge_first.reversed()
            edges.append(edge_first)
            points.append(self.point)
            if child.children:
                child_edges, child_points = child.calc_polygon_edges_and_points(cap=cap,
                                                    reverse_caps=reverse_caps)
                edges.extend(child_edges)
                points.extend(child_points)
            elif cap:
                cap_edge = edge_first.make_cap(child.point)
                if reverse_caps:
                    cap_edge.reverse()
                edges.append(cap_edge)
                points.append(child.point)
            edges.append(edge_last)
            points.append(child.point)
        if cap_start and len(self.children) == 1:
            cap_edge = edge_last.make_cap(self.point)
            if reverse_caps ^ reverse_start_cap:
                cap_edge.reverse()
            edges.append(cap_edge)
            points.append(self.point)
        return edges, points
    def to_drawables(self, draw_verts=False, draw_edges=False,
                     draw_continuous=False, reverse_caps=False, reverse_start_cap=False,
                     **kwargs):
        ds = []
        if draw_continuous:
            edges, points = self.calc_polygon_edges_and_points(cap=True, cap_start=True,
                                        reverse_caps=reverse_caps,
                                        reverse_start_cap=reverse_start_cap)
            poly = Polygon(edges=edges, vertices=points, join=False)
            #poly.offset_polygon()
            ds0 = poly.to_drawables(**kwargs)
            ds.extend(ds0)
        if draw_edges:
            for child in self.children:
                edge = Line.from_points(*self.point, *child.point, segment=True)
                ds0 = edge.to_drawables(**kwargs)
                ds.extend(ds0)
        if draw_verts:
            ds0 = self.point.to_drawables(**kwargs)
            ds.extend(ds0)
        for child in self.children:
            ds.extend(child.to_drawables(draw_verts=draw_verts, draw_edges=draw_edges,
                                         draw_continuous=False, **kwargs))
        return ds
```

```python
def make_hyper_dragon(p1, p2, depth, reverse=False, angle=90, scale=1, fraction=0.5):
    if depth <= 0:
        head = TreeNode(p1)
        tail = TreeNode(p2)
        if reverse:
            head, tail = tail, head
        head.add_child(tail)
        return head, tail
    else:
        d = p1.distance_to(p2)
        trans1 = Transform.translation(Point.from_h_polar(d*fraction,0))
        trans2 = Transform.translation(p1, p2)
        trans = Transform.merge(trans1, trans2)
        p_mid = trans.apply_to_point(Point.from_h_polar(d/2*scale,deg=-angle))
        head1, tail1 = make_hyper_dragon(p_mid, p1, depth-1, reverse=not reverse,
                                       angle=angle, scale=scale, fraction=fraction)
        head2, tail2 = make_hyper_dragon(p_mid, p2, depth-1, reverse=reverse,
                                       angle=angle, scale=scale, fraction=fraction)
        if reverse:
            head1, tail1, head2, tail2 = head2, tail2, head1, tail1
        tail1.add_child(head2.children[0])
        return head1, tail2

def gen_p1_p2():
    global p1, p2
    p1 = Point.from_h_polar(length/2, deg=0)
    p2 = Point.from_h_polar(length/2, deg=-180)

choice = 'mod1'
if choice == 'mod1':
    # Dragon mod1
    depth = 9
    length = 3
    gen_p1_p2()
    dragon, _ = make_hyper_dragon(p1, p2, depth, scale=-1/math.sqrt(3), fraction=0.4)
elif choice == 'mod2':
    # Dragon mod2 (tree, mountain)
    depth = 6
    length = 1
    gen_p1_p2()
    dragon, _ = make_hyper_dragon(p1, p2, depth, scale=.9, fraction=1.1)
else:
    assert False

d = Drawing(2.01, 2.01, origin='center')
d.draw(euclid.Circle(0, 0, 1),
       fill='none', stroke_width=0.006, stroke='black')
line_style = dict(hwidth=0.03, fill='black')
d.draw(dragon, draw_continuous=True,
       hwidth=(-0.002*3,-0.008*3), fill='black', opacity=0.7)

d.set_render_size(w=600)
d.save_svg('images/dragon_{}.svg'.format(choice))
d
```
